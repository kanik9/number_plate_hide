```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as c
```

```python
img = c.imread("Documents/Computer-Vision-with-Python/DATA/car_plate.jpg")
img = c.cvtColor(img,c.COLOR_BGR2RGB)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
display_img(img)
```

```python
plate_cascade = c.CascadeClassifier("/home/kanik/Documents/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_russian_plate_number.xml")
```

```python
def detect_plate(img):
    plate_img = img.copy()
    plate_rects = plate_cascade.detectMultiScale(plate_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in plate_rects:
        c.rectangle(plate_img ,(x,y),(x+w,y+h),(255,0,0),10)
    return plate_img  
```

```python
result = detect_plate(img)
```

```python
plt.imshow(result)
```

```python
blur_img = c.medianBlur(result,5)
display_img(blur_img)
```
